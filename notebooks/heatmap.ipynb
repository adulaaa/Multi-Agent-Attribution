# Agent‑to‑Agent Influence Heatmaps

This notebook loads attribution matrices (dot/cos) from `run_whitebox_debate.py` and plots heatmaps of agent‑to‑agent influences.

```python
import sys
sys.path.append('..')
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

results_dir = "../results"
pt_files = glob.glob(os.path.join(results_dir, "wb-debate-*-attribution.pt"))
if not pt_files:
    print("No attribution .pt files found. Run experiments/run_whitebox_debate.py first.")
else:
    latest = max(pt_files, key=os.path.getctime)
    data = torch.load(latest)
    labels = data["labels"]
    cos = data["cos"].numpy()
    dot = data["dot"].numpy()
    print(f"Loaded {latest}")
    print(f"Labels: {labels}")
```

## Cosine Similarity Heatmap

```python
if 'cos' in locals():
    plt.figure(figsize=(10, 8))
    sns.heatmap(cos, annot=True, fmt=".3f", xticklabels=labels, yticklabels=labels,
                cmap="coolwarm", vmin=-1, vmax=1)
    plt.title("Agent‑to‑Agent Influence (TracIn Cosine Similarity)")
    plt.tight_layout()
    plt.show()
```

## Dot Product Heatmap

```python
if 'dot' in locals():
    plt.figure(figsize=(10, 8))
    sns.heatmap(dot, annot=True, fmt=".1f", xticklabels=labels, yticklabels=labels,
                cmap="viridis")
    plt.title("Agent‑to‑Agent Influence (TracIn Dot Product)")
    plt.tight_layout()
    plt.show()
```

## Top influences on the final turn

```python
if 'cos' in locals():
    last_idx = len(labels)-1
    influence_on_last = cos[last_idx, :]
    top_indices = np.argsort(influence_on_last)[::-1][:3]
    print("Top influences on the final turn:")
    for idx in top_indices:
        print(f"  {labels[idx]}: {influence_on_last[idx]:.4f}")
```
