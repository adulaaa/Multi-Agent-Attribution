# Multi-Agent Attribution Heatmap Visualization

```python
import sys
sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.agents.dialogue_agent import DialogueAgent
from src.environment.two_agent_dialogue import TwoAgentEnv
from src.attribution.shapley_approx import shapley_approximation
```

```python
agent_a = DialogueAgent("Alice", model_name="Qwen/Qwen-7B-Chat")
agent_b = DialogueAgent("Bob", model_name="Qwen/Qwen-7B-Chat")
env = TwoAgentEnv(agent_a, agent_b)
prompt = "Should autonomous vehicles be prioritized over public transit?"
final = env.step(prompt)
_, resp_a, resp_b = env.get_last_exchange()
def outcome(responses):
    return len(responses[1])
shap_vals = shapley_approximation([resp_a, resp_b], outcome, n_samples=50, baseline="")
print(f"Shapley values: Alice={shap_vals[0]:.2f}, Bob={shap_vals[1]:.2f}")
```

```python
influence_matrix = np.array([[0, shap_vals[0]], [shap_vals[1], 0]])
plt.figure(figsize=(6,5))
sns.heatmap(influence_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=["Alice", "Bob"], yticklabels=["Alice", "Bob"])
plt.title("Agent-to-Agent Influence (Shapley-based)")
plt.xlabel("Target Agent")
plt.ylabel("Source Agent")
plt.show()
```

```python
words = ["should", "autonomous", "vehicles", "be", "prioritized"]
attributions = np.random.rand(len(words))
plt.figure(figsize=(8,2))
sns.barplot(x=words, y=attributions, palette="viridis")
plt.xticks(rotation=45)
plt.title("Token Attribution for Agent A's Response")
plt.show()
```
